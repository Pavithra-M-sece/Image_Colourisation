```python
IMG_SIZE = 128  # You can use 64, 128, etc. – tradeoff between quality and Colab RAM

img_dir = './dataset/data/train_color' # Corrected path to the training images
grays = []
colors = []

def load_images(directory):
    img_list = []
    for fname in os.listdir(directory):
        if fname.lower().endswith(('.jpg', '.png', '.jpeg')):
            img = cv2.imread(os.path.join(directory, fname))
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
            img_list.append(img)
    return np.array(img_list)

images = load_images(img_dir)
for img in images:
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    gray = gray.reshape(IMG_SIZE, IMG_SIZE, 1)
    grays.append(gray/255.)
    colors.append(img/255.)

grays = np.array(grays)
colors = np.array(colors)

# Split data
X_train, X_test, y_train, y_test = train_test_split(grays, colors, test_size=0.2, random_state=42)
```

```python
def build_colorization_model():
    input_img = Input(shape=(IMG_SIZE, IMG_SIZE, 1))

    # Encoder: downsampling + feature extraction
    conv1 = Conv2D(64, (3,3), padding='same', activation='relu')(input_img)
    conv1 = BatchNormalization()(conv1)
    conv2 = Conv2D(128, (3,3), padding='same', strides=2, activation='relu')(conv1)
    conv2 = BatchNormalization()(conv2)

    # Bottleneck feature extraction
    conv3 = Conv2D(256, (3,3), padding='same', strides=2, activation='relu')(conv2)
    conv3 = BatchNormalization()(conv3)
    conv4 = Conv2D(512, (3,3), padding='same', activation='relu')(conv3)
    conv4 = BatchNormalization()(conv4)

    # Decoder: upsampling and concatenation
    # Adjusting strides and padding to match conv3 spatial dimensions (32x32)
    up1 = tf.keras.layers.Conv2DTranspose(256, (3, 3), strides=1, padding='same', activation='relu')(conv4) # Keep strides at 1
    up1 = BatchNormalization()(up1)
    up1 = concatenate([up1, conv3])
    conv5 = Conv2D(256, (3,3), padding='same', activation='relu')(up1)
    conv5 = BatchNormalization()(conv5)

    up2 = tf.keras.layers.Conv2DTranspose(128, (3, 3), strides=2, padding='same', activation='relu')(conv5)
    up2 = BatchNormalization()(up2)
    up2 = concatenate([up2, conv2])
    conv6 = Conv2D(128, (3,3), padding='same', activation='relu')(up2)
    conv6 = BatchNormalization()(conv6)

    # Upsample to original image size (128x128)
    up3 = tf.keras.layers.Conv2DTranspose(64, (3, 3), strides=2, padding='same', activation='relu')(conv6)
    up3 = BatchNormalization()(up3)


    # Final color prediction layer
    conv7 = Conv2D(64, (3,3), padding='same', activation='relu')(up3)
    conv7 = BatchNormalization()(conv7)
    output_img = Conv2D(3, (1,1), activation='sigmoid', padding='same')(conv7)

    model = Model(input_img, output_img)
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
                  loss='mse',
                  metrics=['accuracy'])
    return model

model = build_colorization_model()
model.summary()
```

```python
callbacks = [
    tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
    tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3)
]

history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=40,  # You can increase if loss keeps dropping
    batch_size=16,  # 32 may need more RAM
    callbacks=callbacks
)
```
