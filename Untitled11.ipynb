# AI Image Colorization — Proper LAB Colorspace Training

This notebook trains a U-Net on **LAB colorspace**:
- **Input**: L channel (grayscale, 0-100)
- **Output**: ab channels (color, -128 to 127)

Why LAB? RGB+MSE always collapses to gray (average of all colors = gray).  
LAB separates luminance from color, so the model learns ONLY color mapping.

```python
IMG_SIZE = 128
BATCH_SIZE = 32
EPOCHS = 50
MAX_IMAGES = 4000  # Use all 4000 training images
COLOR_FOLDER = './dataset/data/train_color'
```

```python
print('Loading and converting images to LAB colorspace...')

L_images = []   # Input: grayscale (L channel)
ab_images = []  # Target: color (ab channels)

files_list = sorted([f for f in os.listdir(COLOR_FOLDER)
                     if f.lower().endswith(('.jpg', '.jpeg', '.png'))])[:MAX_IMAGES]

for i, fname in enumerate(files_list):
    if i % 500 == 0:
        print(f'  {i}/{len(files_list)}')
    try:
        bgr = cv2.imread(os.path.join(COLOR_FOLDER, fname))
        if bgr is None:
            continue
        bgr = cv2.resize(bgr, (IMG_SIZE, IMG_SIZE))
        # Convert to LAB (float32, L: 0-100, ab: -128 to 127)
        lab = cv2.cvtColor(bgr.astype(np.float32) / 255.0, cv2.COLOR_BGR2Lab)
        # Normalize
        L = lab[:, :, 0:1] / 100.0          # L: 0 to 1
        ab = lab[:, :, 1:] / 128.0          # ab: -1 to 1
        L_images.append(L)
        ab_images.append(ab)
    except Exception as e:
        continue

L_images = np.array(L_images, dtype=np.float32)
ab_images = np.array(ab_images, dtype=np.float32)

print(f'Loaded {len(L_images)} images')
print(f'L shape: {L_images.shape}, ab shape: {ab_images.shape}')
print(f'L range: [{L_images.min():.2f}, {L_images.max():.2f}]')
print(f'ab range: [{ab_images.min():.2f}, {ab_images.max():.2f}]')
```

```python
X_train, X_val, y_train, y_val = train_test_split(
    L_images, ab_images, test_size=0.1, random_state=42
)
print(f'Train: {X_train.shape[0]}, Val: {X_val.shape[0]}')
```

```python
def build_lab_unet(input_shape=(128, 128, 1)):
    """
    U-Net that takes L channel (128x128x1) and predicts ab channels (128x128x2).
    LAB colorspace ensures model learns actual colors, not just brightness.
    """
    inputs = Input(shape=input_shape)

    # Encoder
    def enc_block(x, filters, dropout=0.0):
        x = Conv2D(filters, 3, padding='same')(x)
        x = BatchNormalization()(x)
        x = Activation('relu')(x)
        x = Conv2D(filters, 3, padding='same')(x)
        x = BatchNormalization()(x)
        x = Activation('relu')(x)
        if dropout > 0:
            x = Dropout(dropout)(x)
        return x

    # Encoder path
    e1 = enc_block(inputs, 64)
    p1 = MaxPooling2D(2)(e1)   # 64x64

    e2 = enc_block(p1, 128)
    p2 = MaxPooling2D(2)(e2)   # 32x32

    e3 = enc_block(p2, 256)
    p3 = MaxPooling2D(2)(e3)   # 16x16

    e4 = enc_block(p3, 512, dropout=0.3)
    p4 = MaxPooling2D(2)(e4)   # 8x8

    # Bottleneck
    b = enc_block(p4, 1024, dropout=0.3)  # 8x8

    # Decoder path
    d4 = Conv2DTranspose(512, 2, strides=2, padding='same')(b)
    d4 = concatenate([d4, e4])
    d4 = enc_block(d4, 512)

    d3 = Conv2DTranspose(256, 2, strides=2, padding='same')(d4)
    d3 = concatenate([d3, e3])
    d3 = enc_block(d3, 256)

    d2 = Conv2DTranspose(128, 2, strides=2, padding='same')(d3)
    d2 = concatenate([d2, e2])
    d2 = enc_block(d2, 128)

    d1 = Conv2DTranspose(64, 2, strides=2, padding='same')(d2)
    d1 = concatenate([d1, e1])
    d1 = enc_block(d1, 64)

    # Output: 2 channels for ab, tanh activation (output range -1 to 1)
    outputs = Conv2D(2, 1, activation='tanh', padding='same')(d1)

    model = Model(inputs, outputs)
    return model


model = build_lab_unet()
model.compile(
    optimizer=Adam(learning_rate=1e-3),
    loss='mse',
    metrics=['mae']
)
model.summary()
```

```python
callbacks = [
    EarlyStopping(monitor='val_loss', patience=8, restore_best_weights=True, verbose=1),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=4, min_lr=1e-6, verbose=1),
    ModelCheckpoint('colorization_lab_unet.keras', monitor='val_loss',
                    save_best_only=True, verbose=1)
]

history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    callbacks=callbacks,
    verbose=1
)
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.legend()
plt.title('Loss')
plt.subplot(1, 2, 2)
plt.plot(history.history['mae'], label='Train MAE')
plt.plot(history.history['val_mae'], label='Val MAE')
plt.legend()
plt.title('MAE')
plt.tight_layout()
plt.savefig('training_history.png', dpi=100)
plt.show()
```

```python
def colorize_lab(model, L_input):
    """Colorize a single L channel image using the trained model."""
    L_batch = np.expand_dims(L_input, axis=0)
    ab_pred = model.predict(L_batch, verbose=0)[0]

    # Denormalize
    L_denorm = L_input[:, :, 0] * 100.0
    ab_denorm = ab_pred * 128.0

    lab_out = np.zeros((IMG_SIZE, IMG_SIZE, 3), dtype=np.float32)
    lab_out[:, :, 0] = L_denorm
    lab_out[:, :, 1:] = ab_denorm

    bgr = cv2.cvtColor(lab_out, cv2.COLOR_Lab2BGR)
    rgb = cv2.cvtColor((np.clip(bgr, 0, 1) * 255).astype(np.uint8), cv2.COLOR_BGR2RGB)
    return rgb

# Show 5 test samples
n = 5
indices = np.random.choice(len(X_val), n, replace=False)

fig, axes = plt.subplots(n, 3, figsize=(12, n * 4))
for i, idx in enumerate(indices):
    # Grayscale input
    axes[i, 0].imshow(X_val[idx, :, :, 0], cmap='gray')
    axes[i, 0].set_title('Grayscale Input')
    axes[i, 0].axis('off')

    # Predicted colorization
    pred_rgb = colorize_lab(model, X_val[idx])
    axes[i, 1].imshow(pred_rgb)
    axes[i, 1].set_title('AI Colorized (LAB)')
    axes[i, 1].axis('off')

    # Ground truth
    L_denorm = X_val[idx, :, :, 0] * 100.0
    ab_denorm = y_val[idx] * 128.0
    lab_true = np.stack([L_denorm, ab_denorm[:, :, 0], ab_denorm[:, :, 1]], axis=-1).astype(np.float32)
    bgr_true = cv2.cvtColor(lab_true, cv2.COLOR_Lab2BGR)
    rgb_true = cv2.cvtColor((np.clip(bgr_true, 0, 1) * 255).astype(np.uint8), cv2.COLOR_BGR2RGB)
    axes[i, 2].imshow(rgb_true)
    axes[i, 2].set_title('Ground Truth')
    axes[i, 2].axis('off')

plt.tight_layout()
plt.savefig('colorization_results.png', dpi=100, bbox_inches='tight')
plt.show()
print('Results saved!')
```
